## 1. Obtenção dos Dados

Primeiramente, começa-se por importar as bibliotecas necessárias para o desenvolvimento do trabalho. 

A biblioteca __Pandas__ facilita a análise e manipulação de dados. Possui uma estrutura de dados chave designada por _data frame_ que se assemelha a uma tabela de Excel ,por exemplo.

Por sua vez a biblioteca __Numpy__ é fundamental na ciência computacional. Possui um tipo de dados designado po _ndarray_ que se assemelha a uma lista, em que todos os elementos da mesma tẽm de ser obrigatoriamente do mesmo tipo, facilitando também a aplicação de operações matemáticas e de álgebra.

A biblioteca __Matplotlib__ que permite uma melhor visualização de dados através de gráficos.

Por fim a biblioteca __Scikit-Lear__ oferece ferramentas para para _machine learning_, incluindo algoritmos de classificação, regressão e agrupamento (mais conhecido por _clustering_ ).

```python
# Import the necessary packages
import pandas as pd
import numpy as np
import matplotlib as mp
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_predict
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier


# Read the data from CSV files
olympics_data = pd.read_csv('athlete_events.csv')
country_data = pd.read_csv('noc_regions.csv')

# Printing the first 5 rows of the athletes data
olympics_data.head() 
```

```python
# Printing the first 5 rows of the countries data
country_data.head() 
```

```python
# Printing some informations about the athletes data
olympics_data.info()
```

```python
# Printing some informations about the countries data
country_data.info()
```

Como se pode observar acima, existem algumas colunas (_Age, Heigh, Weight_ e _Medal_ ) que não tẽm as mesmas quantidades de dados que as restantes. No caso da coluna _Medal_ é espectável que tenha um número inferior de dados, visto que nem todos os atletas participantes acabam por ser medalhados. No que toca às restantes, talvez seja necessário fazer uma limpeza dos dados. Para tal é necessário saber quantos dados estão em falta em cada coluna:

```python
# Printing the missing values in each column
print(olympics_data.isnull().sum())
```

## 2. Tratamento e Limpeza dos Dados

Depois de se saber a quantidade de dados em falta em cada coluna pode-se então passar ao tratamento. Nesse sentido optou-se por pôr na coluna _Medal_ o valor "DNW" (Did Not Win, em português, não ganhou), nos dados em falta. 

```python
# Replacing the NaN value with 'DNW'
olympics_data['Medal'].fillna('DNW', inplace = True)

# Printing the missing values in each column
print(olympics_data.isnull().sum())
```

Em relação aos dados em falta nas restantes colunas, optou-se por apagar esses dados de modo a ter um _dataset_ o mais íntegro possível.

```python
# Deleting the rows that contain null values in the Age, Height and Weight columns.
olympics_data.dropna(axis = 0, subset=['Age', 'Height', 'Weight'], inplace = True)
olympics_data.reset_index()
print(olympics_data.isnull().sum(), '\n')
olympics_data.info()
```

Pode-se agora observar que o _dataset_ já não contém linhas com valores nulos. Em seguida fundiu-se os dados dos atletas com os países de modo a obter um único _dataframe_. 

```python
# Rename the column "region" and drop the column 'Notes'
country_data.rename(columns = {'region' : 'Country'}, inplace = True)
country_data.drop('notes', axis = 1, inplace = True)

# Merge the dataframes
olympics_data = pd.merge(olympics_data, country_data, on='NOC', how='left')
olympics_data.head() 
```

Por fim resta eliminar eventuais dados duplicados.

```python
olympics_data.drop_duplicates(inplace = True)
```

## 3. Análise dos Dados 

Nesta secção procede-se a uma análise dos dados de modo obter algumas informações úteis sobre os mesmos. 

Uma forma de se observar o tipo de dados é através de histogramas dos atributos numéricos:

```python
olympics_data.hist(bins=50, figsize=(20,20))
mp.pyplot.show()
```

Tendo em conta o problema que se pretende resolver, as variáveis mais significativas são a idade, a altura, o peso, o desporto e o género. Nesse sentido criou-se um _dataset_ secundundário que contempla apenas estas variáveis.

```python
# Deleting the unnecessary columns
olympics_sub_data = olympics_data.drop(columns = ['ID','Name','Team', 'Games', 'Season', 'City', 'NOC', 'Event', 'Medal', 'Country', 'Year'])

# Transforming the column sex and sport into numerical values
sex_encoder = LabelEncoder()
sport_encoder = LabelEncoder()
olympics_sub_data['Sex'] = sex_encoder.fit_transform(olympics_sub_data['Sex'])
olympics_sub_data['Sport'] = sport_encoder.fit_transform(olympics_sub_data['Sport'])
olympics_sub_data.info()
```

Pode-se observar acima que no _dataset_ secundário, os dados encontram-se todos com valores numéricos,facilitando assim a análise de correlações entre os mesmos. 

```python
# Correlation matrix between 'Sport' and other variables
corr_matrix = olympics_sub_data.corr()
corr_matrix['Sport'].sort_values()
```

Observando os resultados obtidos, deduz-se que as restantes variáveis não apresentam grande correlação com o Desporto, dado que os respetivos valores estão muito pŕoximos de 0. Outra forma de visualizar a correlação das variáveis é utilizar a função _scatter_matrix_. 

```python
# Plotting the correlation between the variables
pd.plotting.scatter_matrix(olympics_sub_data, alpha=0.5, figsize=(20,20))
```

Observando os gráficos acima pode-se concluir, mais uma vez, que as variáveis não apresentam grande correlação com a variável desporto. No entanto é verificável que existe um correlação positiva entre a altura e o peso, o que é expectável. Dado que não existe nenhuma correlação forte entre as variáveis que se pretendem usar, não é expectável que algum modelo seja muito acertivo no que toca à classificação.

## 4. Seleção do Modelo de _Machine Learning_ 

De modo a selecionar o melhor modelo para resolver este problema, foi utilizado o seguinte mapa: 
![Logo](https://scikit-learn.org/stable/_static/ml_map.png)
Uma vez que se pretende classificar o desporto que um certo atleta pratica consoante a sua idade, altura e peso, estamos perante um problema de "_supervised learning_ "  em que os dados utilizados para treinar o modelo estão classificados, ou seja, no ato da classificação o modelo conhece _a priori_ todas as classes.

Nesse sentido, seguindo o mapa anterior, é sugerido que se utilize o __SGD Classifier__, que está na área dos problemas de classificação. Consiste numa otimização dos modelos _Support Vector Machine_ e _Logistic Regression_ (_SGD Regressor_ neste caso). Este modelo seleciona uma instância aleatória no conjunto de dados de treino e em cada etapa calcula os gradientes com base nessa única instância. Quando o conjunto de dados é relativamente grande, a utilização do _SVM_ , que utiliza todo o conjunto de dados de treino, torna-se muito dispendiosa tanto nos recursos computacionais, como no tempo de execução. Por outro lado o _SGD_ apresenta uma alta escalabilidade tornando-se assim uma melhor escolha para grandes conjuntos de dados.

Na [documentação](https://scikit-learn.org/stable/modules/sgd.html#tips-on-practical-use) é sugerido que antes de se fazer o treino deste modelo se faça a normalização dos dados. Para tal foi utilizado um _pipeline_ que, de uma forma resumida, primeiro utiliza o _StandardScaler_ para normalizar os dados e só depois aplica o modelo.

```python
# Dividing the test data into train and test sets 
x = olympics_sub_data.drop(columns='Sport')
y = olympics_sub_data['Sport']
x_train, x_test, y_train, y_test = train_test_split(x.values, y.values, test_size=0.3) 

# Defining the model
model = make_pipeline(StandardScaler(), SGDClassifier(loss="log"))

# Trainning the model
model.fit(x_train, y_train)

# Make predictions on the same set used to train
predictions = model.predict(x_train)
print("Dados de treino:\n", classification_report(y_train, predictions), "\n")

# Make predictions on the test set
predictions = model.predict(x_test)
print("Dados de teste:\n", classification_report(y_test, predictions))
```

Observando os resultados acima, conclui-se que este modelo tem uma precisão de apenas 15% quer nos dados para treino, quer nos dados para teste. Apesar de ter sido recomendado pelo mapa referido anteriormente, o modelo _SGD_ é um modelo do tipo binário o que significa que funciona melhor quando só existem duas classes para serem atribuidas. Experimentando um modelo modelo multiclasse, espera-se que os resulados sejam melhores.

```python
# Defining the model
model = make_pipeline(StandardScaler(), RandomForestClassifier())

# Trainning the model
model.fit(x_train, y_train)

# Make predictions on the same set used to train
predictions = model.predict(x_train)
print("Dados de treino:\n", classification_report(y_train, predictions), "\n")

# Make predictions on the same set used to train
predictions = model.predict(x_test)
print("Dados de teste:\n", classification_report(y_test, predictions))
```

Após experimentar o modelo multiclasse _Ramdom Forest Classifier_ , que consiste em criar subsets de dados e usá-los para treinar modelos de _Decision Tree_ , verifica-se que a precisão nos dados de treino aumentou para 53% e nos dados de teste para 28%. Portanto, este modelo representa uma melhor solução que o anterior. 

O próximo passo para melhorar a precisão do modelo seria a sintonização dos hiperparâmetros. Este procedimento consiste em experimentar diferentes combinações de parâmetros do respetivo modelo de modo a encontrar a combinação que resulta na maior precisão possível. A lista de parâmetros do modelo _Ramdom Forest Classifier_ pode ser encontrada na [documentação](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html?highlight=random%20forest#sklearn.ensemble.RandomForestClassifier).

## 5. Classificação
